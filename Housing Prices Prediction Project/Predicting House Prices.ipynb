Outliers can severely affect the later models ability to generalise well to new data. An outlier will therefore sit way outside of the distribution of data points. Hence, this will skew the distribution of the data and potential calculations. ### 2. Fill Missing Data

If you have missing values, you have two options: 1) Delete the entire row 2) Fill the missing entry with an imputed value

```python
# visualise the missing data
fig, axes = plt.subplots(1,2,figsize=(35,10))
sns.heatmap(train.isnull(), cbar=False, ax=axes[0], cmap="YlGnBu_r").set_title("Training Dataset")
sns.heatmap(test.isnull(), cbar=False, ax=axes[1], cmap="YlGnBu_r").set_title("Test Dataset");
```

```python
# Using data description, fill these missing values with "None"
for col in ("PoolQC", "MiscFeature", "Alley", "Fence", "FireplaceQu",
           "GarageType", "GarageFinish", "GarageQual", "GarageCond",
           "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1",
            "BsmtFinType2", "MSSubClass", "MasVnrType"):
    train[col] = train[col].fillna("None")
    test[col] = test[col].fillna("None")
```

```python
# The area of the lot out front is likely to be similar to the houses in the local neighbourhood
train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
test["LotFrontage"] = test.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
# Using data description, fill these missing values with 0 
for col in ("GarageYrBlt", "GarageArea", "GarageCars", "BsmtFinSF1", 
           "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "MasVnrArea",
           "BsmtFullBath", "BsmtHalfBath"):
    train[col] = train[col].fillna(0)
    test[col] = test[col].fillna(0)
```

```python
# Fill these features with their mode, the most commonly occuring value. This is okay since there are a low number of missing values for these features
test['Electrical'] = test['Electrical'].fillna(test['Electrical'].mode()[0])
test['KitchenQual'] = test['KitchenQual'].fillna(test['KitchenQual'].mode()[0])
test['Exterior1st'] = test['Exterior1st'].fillna(test['Exterior1st'].mode()[0])
test['Exterior2nd'] = test['Exterior2nd'].fillna(test['Exterior2nd'].mode()[0])
test['SaleType'] = test['SaleType'].fillna(test['SaleType'].mode()[0])
test["Functional"] = test["Functional"].fillna(test['Functional'].mode()[0])
```

```python
# fill GarageType with 'Detchd' as imputed by the mean
train[train.GarageType.isnull()].OverallQual.mean()
train['GarageType'] = train['GarageType'].fillna("Detchd")
train['GarageType'].shape
```

```python
test[test.GarageType.isnull()].OverallQual.mean()
test['GarageType'] = test['GarageType'].fillna("Detchd")
test['GarageType'].shape
```

```python
# Fill MSZoning with 'RL' - most common based on supporting variables
test[test['MSZoning'].isnull()]
test['MSZoning'] = test['MSZoning'].fillna("RL")
test['MSZoning'].shape
```

```python
# Fill Total Basement SF with the mean SF
test['TotalBsmtSF'] = test['TotalBsmtSF'].fillna(test['TotalBsmtSF'].mean())
test['TotalBsmtSF'].shape
```

```python
# Drop Utilities columns
train = train.drop(columns='Utilities')
test = test.drop(columns='Utilities')
```

```python
# Drop Id columns
train = train.drop(columns='Id')
test = test.drop(columns='Id')
```

### 4. Feature Engineering

I returned to this step after intial modelling, to improve the model by engineering further features. Highly correlated features will be simplified and built into <a href="https://nycdatascience.com/blog/student-works/predicting-housing-prices-in-ames-iowa-2/"> new variables.</a> 
<li>Remodeled: Binary value representing if the has remodeled year was different than house built year</li>
<li>TotalSqFeet: Add all square feet data together.</li>
<li>TotalPorchSF: Add all square feet data together.</li>
<li>House Age: The difference between Year Sold and Year Remodeled</li>
<li>Is New: If Year Sold equals Year Built</li>

#### Total Square Feet

```python
# Add total square feet
train['TotalSqFeet'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF'] 
test['TotalSqFeet'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']
        
train = train.drop(columns=['TotalBsmtSF', '1stFlrSF', '2ndFlrSF'])
test = test.drop(columns=['TotalBsmtSF', '1stFlrSF', '2ndFlrSF'])
```

```python
test['TotalSqFeet'] = test['TotalSqFeet'].fillna(0)
```

```python
# set type to integer
test['TotalSqFeet'] = test['TotalSqFeet'].astype('int64')
```

#### Total Bathrooms

```python
# Add total bathrooms
train['TotalBathrooms'] = train['FullBath'] + train['HalfBath'] + train['BsmtFullBath'] + train['BsmtHalfBath']
test['TotalBathrooms'] = test['FullBath'] + test['HalfBath'] + test['BsmtFullBath'] + test['BsmtHalfBath']

train = train.drop(columns=['FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath'])
test = test.drop(columns=['FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath'])
```

#### Remodeled vs Year Built

```python
# Add if the has remodeled year was different than house built year
le = LabelEncoder()
train['Remodeled'] = train['YearRemodAdd'] != train['YearBuilt']
train['Remodeled'] = le.fit_transform(train['Remodeled'])

test['Remodeled'] = test['YearRemodAdd'] != test['YearBuilt']
test['Remodeled'] = le.fit_transform(test['Remodeled'])
```

#### Is New

```python
# Add if the house is new
train['IsNew'] = train['YrSold'] == train['YearBuilt']
train['IsNew'] = le.fit_transform(train['IsNew'])

test['IsNew'] = test['YrSold'] == test['YearBuilt']
test['IsNew'] = le.fit_transform(test['IsNew'])
```

#### House Age

```python
# Add difference between Year Sold and Year Remodeled
train['HouseAge'] = train['YrSold'] - train['YearRemodAdd']
test['HouseAge'] = test['YrSold'] - test['YearRemodAdd']
```

Standardize features by removing the mean and scaling to unit variance with Z-Score (Standard Scaling) or Min-Max (0-1) if distribution is not normal. <a href="http://benalexkeen.com/feature-scaling-with-scikit-learn/">Scaling</a> is important when numerical features have different ranges to avoid bias.

## Gradient Boosting Regressor Model

Gradient boosting is a machine learning technique for regression and classification problems, which produces a prediction model in the form of an ensemble of weak prediction models, typically decision trees. (Wikipedia definition) "The idea is to use the weak learning method several times to get a succession of hypotheses, each one refocused on the examples that the previous ones found difficult and misclassified." Photo from <a href="https://medium.com/mlreview/gradient-boosting-from-scratch-1e317ae4587d">Medium.com</a>

<img src="https://cdn-images-1.medium.com/max/880/1*8T4HEjzHto_V8PrEFLkd9A.png" height="40%" width="40%">

```python
from sklearn.ensemble import GradientBoostingRegressor

# fit and train the gradient boosting regressor using the training set
GBR =  GradientBoostingRegressor(alpha=0.9, criterion='friedman_mse', init=None,
             learning_rate=0.05, loss='huber', max_depth=4,
             max_features='sqrt', max_leaf_nodes=None,
             min_impurity_decrease=0.0, min_impurity_split=None,
             min_samples_leaf=15, min_samples_split=10,
             min_weight_fraction_leaf=0.0, n_estimators=3000,
             presort='auto', random_state=None, subsample=1.0, verbose=0,
             warm_start=False)

GBR =  GBR.fit(X_train, y_train)

# score - returns the coefficient of determination R^2 of the prediction.
print(GBR.score(X_train, y_train))
```

## Cross Validation <a id="CrossValidation">

A solution to this problem is a procedure called cross-validation (CV for short). A test set should still be held out for final evaluation, but the validation set is no longer needed when doing CV. In the basic approach, called k-fold CV, the training set is split into k smaller sets (other approaches are described below, but generally follow the same principles). The following procedure is followed for each of the k “folds”:

A model is trained using  of the folds as training data;
the resulting model is validated on the remaining part of the data (i.e., it is used as a test set to compute a performance measure such as accuracy). Information and diagram from <a href="https://scikit-learn.org/stable/modules/cross_validation.html">documentation.</a>

<img src="https://scikit-learn.org/stable/_images/grid_search_cross_validation.png" height="35%" width="35%">
