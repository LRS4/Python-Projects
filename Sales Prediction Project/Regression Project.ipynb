### Model Preparation

You have explored and analyzed customer data collected by the Adventure Works Cycles company. Now you should be ready to apply what you have learned about the data to building, testing, and optimizing a predictive machine learning model.

First I will check the test dataset.

```python
train = data
test =  pd.read_csv("AW_test.csv")
```

```python
train.shape, test.shape
```

```python
# drop columns with missing data
test.drop(columns=["Title", "MiddleName", "Suffix", "AddressLine2"], inplace=True)
```

```python
# check for all duplicate rows
test.duplicated().sum()
```

```python
test.isnull().sum()
```

```python
train.shape, test.shape
```

```python
# encode gender and marital status
var_mod = ['Gender', 'MaritalStatus', 'Education', 'Occupation', 'City', 'StateProvinceName', 'CountryRegionName']
le = LabelEncoder()
for i in var_mod:
    train[i] = le.fit_transform(train[i])
```

```python
# encode gender and marital status
var_mod = ['Gender', 'MaritalStatus', 'Education', 'Occupation', 'City', 'StateProvinceName', 'CountryRegionName']
le = LabelEncoder()
for i in var_mod:
    test[i] = le.fit_transform(test[i])
```

```python
train.head(1)
```

```python
test.head(1)
```

### Train / Test Split

```python
from sklearn.model_selection import train_test_split

# target
y = train.pop('AveMonthSpend').values
```

```python
# training features
X = train.select_dtypes('number').drop(columns=['CustomerID', 'Age', 'BikeBuyer']).values
```

```python
# test features
test_features = test.select_dtypes('number').drop(columns=['CustomerID']).values
```

```python
X.shape, test_features.shape
```

```python
# split the training set into train and test splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
X_train.shape, X_test.shape
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

# fit the scaler based upon the training set
scaler = StandardScaler().fit(X_train)

# scale the train features array 
X_train = scaler.transform(X_train)

# scale the test features array (based on the training fit)
X_test = scaler.transform(X_test)
```

```python
# ensure target average monthly spend is converted to log values
y_train = np.log(y_train)
y_test = np.log(y_test)
```

```python
print(y_train), print(y_test)
```

### Modelling

```python
import sklearn.metrics as sklm
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# build and fit the random_forest classifer using the training set
gbr = GradientBoostingRegressor()

gbr = gbr.fit(X_train, y_train)

# print the score for the training set
print(gbr.score(X_train, y_train))
```

```python
# apply the model to the test set and output score - held back data
gbr.score(X_test, y_test)
```

```python
y_pred = gbr.predict(X_test)

## Print the usual metrics and the R^2 values
print('Mean Square Error      = ' + str(sklm.mean_squared_error(y_test, y_pred)))
print('Root Mean Square Error = ' + str(math.sqrt(sklm.mean_squared_error(y_test, y_pred))))
print('Mean Absolute Error    = ' + str(sklm.mean_absolute_error(y_test, y_pred)))
print('Median Absolute Error  = ' + str(sklm.median_absolute_error(y_test, y_pred)))
print('R^2                    = ' + str(sklm.r2_score(y_test, y_pred)))
```

```python
# predicting on the training set
train_y_pred = gbr.predict(X_train)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].set_title("Distribution of residuals")

## make the residual plots - should result centred on 0
sns.distplot(y_test - y_pred, bins=10, ax=ax[0]);

## now make the residual plots - models train vs test predictions
sns.regplot(y = train_y_pred, x = y_train, color = 'blue', label = 'Training Data', scatter_kws={'alpha':0.75});
sns.regplot(y = y_pred, x = y_test, color = 'green', label = 'Validation Data', scatter_kws={'alpha':0.75});
plt.title('Predicted Values vs Test Values');
plt.xlabel('Real Values');
plt.ylabel('Predicted Values');
plt.legend(loc = 'upper left');
```

```python
# change target and predicts back to prices from logarithmic scale

fig, ax = plt.subplots(1,figsize=(14,5))
sns.regplot(y = np.exp(train_y_pred), x = np.exp(y_train), color = 'blue', label = 'Training Data', scatter_kws={'alpha':0.75});
sns.regplot(y = np.exp(y_pred), x = np.exp(y_test), color = 'green', label = 'Validation Data', scatter_kws={'alpha':0.75});
plt.title('Gradient Boosting Regressor - Predicted Values vs Test Values');
plt.xlabel('Real Values');
plt.ylabel('Predicted Values');
plt.legend(loc = 'upper left');
```

```python
# pass all the data to cross validation, as k-folds will split the data
from sklearn.model_selection import cross_val_score
cv_gbr_scores = cross_val_score(gbr, X, y, cv=10)
print(cv_gbr_scores)
print("Accuracy: " + str(round(cv_gbr_scores.mean() * 100,2)) + "%")
print("Variance: " + str(round(cv_gbr_scores.std() * 100,2)) + "% (+/-)")
```

```python
feature_importances = pd.Series(gbr.feature_importances_, index=['City', 'StateProvinceName', 'CountryRegionName', 'Education', 'Occupation', 'Gender','MaritalStatus','HomeOwnerFlag','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','YearlyIncome'])
feature_importances.sort_values(ascending=True).plot(kind="barh", edgecolor="black", color="#1F77B4", figsize=(12,5));
```
